```python
df_review = df.sort_values(by= "Number of google review in lakhs", ascending= False)
df_review
```

```python
df_morning = df_review[df["Best Time to visit"] == "Morning"]
df_morning
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.barplot(x = "Name", y = "Number of google review in lakhs", data = df_morning, ax=ax)
plt.xticks(rotation=90)
plt.show()
```

```python
df_afternoon = df_review[df["Best Time to visit"] == "Afternoon"]
df_afternoon
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.barplot(x = "Name", y = "Number of google review in lakhs", data = df_afternoon, ax=ax)
plt.xticks(rotation=90)
plt.show()
```
